# Working with 2D arrays

1. Working with 2D arrays
Now, let's see how to plot 2D arrays (sometimes called raster data) that represent either images or functions of two variables, also known as bivariate functions. 2. Reminder: NumPy arrays
Let's review some NumPy first. NumPy arrays resemble lists but are uniform in data type. They support vectorized computations or calculations over the entire array without writing loops. Square brackets can index arrays. For 2D arrays, the syntax for indexing is [index0, index1].

3. Reminder: slicing arrays
A slice is a block of array entries. The principle notation for slicing uses colons. When slicing an array, the syntax is start:stop:stride. Remember, slices includes the start index but not the stop index. The Stride tells as at which intervals to slice the array, for example, a stride of 2 will sample every second entry. Also, with negative indices or strides, we count backward from the end of the array.

4. 2D arrays & images

Any computerized black-and-white image is a numerical matrix with entries corresponding to pixel intensities. # Generating meshes
In order to visualize two-dimensional arrays of data, it is necessary to understand how to generate and manipulate 2-D arrays. Many Matplotlib plots support arrays as input and in particular, they support NumPy arrays. The NumPy library is the most widely-supported means for supporting numeric arrays in Python.

In this exercise, you will use the meshgrid function in NumPy to generate 2-D arrays which you will then visualize using `plt.imshow()`. The simplest way to generate a meshgrid is as follows:

```
import numpy as np 
Y,X = np.meshgrid(range(10),range(20))
```

This will create two arrays with a shape of `(20,10)`, which corresponds to 20 rows along the Y-axis and 10 columns along the X-axis. In this exercise, you will use `np.meshgrid()` to generate a regular 2-D sampling of a mathematical function.

Instructions

1. Import the numpy and matplotlib.pyplot modules using the respective aliases np and plt.

2. Generate two one-dimensional arrays u and v using np.linspace(). The array u should contain 41 values uniformly spaced between -2 and +2. The array v should contain 21 values uniformly spaced between -1 and +1.

3. Construct two two-dimensional arrays X and Y from u and v using np.meshgrid().

4. After the array Z is computed using X and Y, visualize the array Z using plt.pcolor() and plt.show().

5. Save the resulting figure as 'sine_mesh.png'.

````
# Import numpy and matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt

# Generate two 1-D arrays: u, v
u = np.linspace(-2, 2, 41)
v = np.linspace(-1, 1, 21)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid(u, v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor()
plt.pcolor(Z)
plt.show()

# Save the figure to 'sine_mesh.png'
plt.savefig('sine_mesh.png')
````

2. Distributions of 2D points

3. Histograms in 1D
The fundamental tool for one-dimensional data is a histogram. That is, given a collection of realizations of a single random variable, we choose a set of intervals or bins. We then count the number of data points that lie in each bin (a process called binning) A histogram consists of the counts or frequencies are plotted over each bin, which can be plotted as shown here.

4. Histograms in 1D
he Matplotlib function hist() can be used to plot histograms like the one just shown. We have the option to choose the number of bins. In one dimension, straight-line segments are the only possible shape for bins in a histogram. For data in two dimensions, bins can be more general shapes. We'll consider two kinds: rectangles & hexagons. The most obvious strategy uses rectangular bins to build a 2D histogram.

# Using hist2d()

Given a set of ordered pairs describing data points, you can count the number of points with similar values to construct a two-dimensional histogram. This is similar to a one-dimensional histogram, but it describes the joint variation of two random variables rather than just one.

In matplotlib, one function to visualize 2-D histograms is plt.hist2d().

 - You specify the coordinates of the points using plt.hist2d(x,y) assuming x and y are two vectors of the same length.

 - You can specify the number of bins with the argument bins=(nx, ny) where nx is the number of bins to use in the horizontal direction and ny is the number of bins to use in the vertical direction.

 - You can specify the rectangular region in which the samples are counted in constructing the 2D histogram. The optional parameter required is `range=((xmin, xmax), (ymin, ymax))` where

 - xmin and xmax are the respective lower and upper limits for the variables on the x-axis and

 - ymin and ymax are the respective lower and upper limits for the variables on the y-axis. Notice that the optional range argument can use nested tuples or lists.

In this exercise, you'll use some data from the auto-mpg data set. There are two arrays mpg and hp that respectively contain miles per gallon and horse power ratings from over three hundred automobiles built.

Instructions

 - Put hp along the horizontal axis and mpg along the vertical axis.

 - Specify 20 by 20 rectangular bins with the bins argument.

 - Specify the region covered by using the optional range argument so that the plot samples hp between 40 and 235 on the x-axis and mpg between 8 and 48 on the y-axis. Your argument should take the form: `range=((xmin, xmax), (ymin, ymax))`.

2. Add a color bar to the histogram.

````
# Generate a 2-D histogram
plt.hist2d(hp, mpg, bins=(20,20), range=((40, 235), range=(8, 48)))
# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hist2d() plot')
plt.show()
````

# Using hexbin()

The function plt.hist2d() uses rectangular bins to construct a two dimensional histogram. As an alternative, the function plt.hexbin() uses hexagonal bins. The underlying algorithm (based on this [article from 1987](https://amstat.tandfonline.com/doi/abs/10.1080/01621459.1987.10478445#.YAu1POhKhPY)) constructs a hexagonal tesselation of a planar region and aggregates points inside hexagonal bins.

 - The optional gridsize argument (default 100) gives the number of hexagons across the x-direction used in the hexagonal tiling. If specified as a list or a tuple of length two, gridsize fixes the number of hexagon in the x- and y-directions respectively in the tiling.

 - The optional parameter `extent=(xmin, xmax, ymin, ymax)` specifies rectangular region covered by the hexagonal tiling. In that case, xmin and xmax are the respective lower and upper limits for the variables on the x-axis and ymin and ymax are the respective lower and upper limits for the variables on the y-axis.
In this exercise, you'll use the same auto-mpg data as in the last exercise (again using arrays mpg and hp). This time, you'll use plt.hexbin() to visualize the two-dimensional histogram.

 - Specify a hexagonal tesselation with 15 hexagons across the x-direction and 12 hexagons across the y-direction using gridsize.

 - Specify the rectangular region covered with the optional extent argument: use hp from 40 to 235 and mpg from 8 to 48.

# Working with images

1. Working with images
We've seen tools for plotting smooth functions and distributions. Let's focus now on actual images.

2. Images
Remember, an image is a matrix of intensity values. That's simple to think of with black-and-white images with values from 0 through 1 representing black through white with shades of gray in between. RGB color images are stored as 3D arrays: one for each channel, red, green, and blue. The values are often floating-point numbers between zero and one or integers between zero and 255.

3. Loading images
Two of the principle commands for working with images are imread() which loads images, and imshow() which displays them. Here, we load in image sunflower(dot)jpg using imread(). The image is loaded into a 3D NumPy array where the trailing dimension - the last to appear in img(dot)shape - is 3. We also use axis('off') to hide axes when displaying images. To display images, we still need to use the Matplotlib show() command to render the image on the screen. This is the default image of the sunflower.

4. Reduction to gray-scale image
There are a number of ways to convert a color image to a grayscale image. A simple one is to average the RGB channels, that is, apply the NumPy mean() method along axis 2 (the trailing dimension). That is, since the color image consists of three matrices with entries between 0 and 255, using the average works as a reasonable proxy for the RGB values when collapsing them to a single scalar intensity value for each pixel. Finding the matrix that is the elementwise average of all three yields a matrix still with entries between 0 and 255. Notice we explicitly set the colormap 'gray' so that subsequent images will all be grayscale. The resulting collapsed array is still 2-dimensional and can be plotted as a 2D grayscale image easily.

5. Uneven samples
One subtlety regarding the imshow() command is pixel shape. Suppose we take the pixels from every 4th row and every 2nd column of the original grayscale image. We can do that in NumPy using the slices; remember that the syntax colon-colon-4 refers to every 4th element and Thus, our subsampled image uneven has dimensions 120 by 320, while the original image was 480 by 640. When plotted using imshow(), the image displayed is distorted. The reason for the distortion is that the default assumption is that pixels are squares of equal width and height. When displaying an unevenly subsampled image, the correct assumption here is that the aspect ratio (the ratio of pixel height over pixel width) is two.

6. Adjusting aspect ratio
The keyword here is aspect. Using aspect equals 2-point-0 in the call to imshow() results in an image with appropriately scaled pixels.

7. Adjusting extent
Another way to fix this problem is to specify the original image dimensions in the optional keyword argument extent. This can be confusing; the order of arguments for extent is from left to right & from bottom to top. In this case, the optional argument is extent=(0,640,0,480). Using the keyword argument extent with an appropriate tuple allows the computation of the correct pixel aspect ratio and, again, a suitable plot results.

# Extent and aspect

When using plt.imshow() to display an array, the default behavior is to keep pixels square so that the height to width ratio of the output matches the ratio determined by the shape of the array. In addition, by default, the x- and y-axes are labeled by the number of samples in each direction.

The ratio of the displayed width to height is known as the image aspect and the range used to label the x- and y-axes is known as the image extent. The default aspect value of 'auto' keeps the pixels square and the extents are automatically computed from the shape of the array if not specified otherwise.

# Rescaling pixel intensities
Sometimes, low contrast images can be improved by rescaling their intensities. 

For instance, this image of Hawkes Bay, New Zealand has no pixel values near 0 or near 255 (the limits of valid intensities). (originally by Phillip Capper, modified by User:Konstable, via Wikimedia Commons, [CC BY 2.0](https://creativecommons.org/licenses/by/2.0/)) 

Hawkes Bay, New Zealand

image.png

For this exercise, you will do a simple rescaling (remember, an image is NumPy array) to translate and stretch the pixel intensities so that the intensities of the new image fill the range from 0 to 255.

Instructions

1. Use the methods .min() and .max() to save the minimum and maximum values from the array image as pmin and pmax respectively.

2. Create a new 2-D array rescaled_image using 256*(image-pmin)/(pmax-pmin)

3. Plot the new array rescaled_image.

````
# Load the image into an array: image
image = plt.imread('640px-Unequalized_Hawkes_Bay_NZ.jpg')

# Extract minimum and maximum values from the image: pmin, pmax
pmin, pmax = image.min(), image.max()
print("The smallest & largest pixel intensities are %d & %d." % (pmin, pmax))

# Rescale the pixels: rescaled_image
rescaled_image = 256*(image - pmin) / (pmax - pmin)
print("The rescaled smallest & largest pixel intensities are %.1f & %.1f." % 
      (rescaled_image.min(), rescaled_image.max()))

# Display the rescaled image
plt.title('rescaled image')
plt.axis('off')
plt.imshow(rescaled_image)

plt.show()

'''
<script.py> output:
    The smallest & largest pixel intensities are 104 & 230.
    The rescaled smallest & largest pixel intensities are 0.0 & 256.0.
'''
````
